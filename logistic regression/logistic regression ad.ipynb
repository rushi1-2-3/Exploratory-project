# To create a model that will predict whether or not they will click on an ad based on the features given.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ad = pd.read_csv('advertising.csv')
```

```python
ad.head()
```

```python
ad.info()
```

```python
ad.describe()
```

```python
sns.set_style('whitegrid')
ad['Age'].hist(bins=30)
plt.xlabel('Age')
```

```python
sns.pairplot(ad,hue="Clicked on Ad")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = ad[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = ad['Clicked on Ad']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions = logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
